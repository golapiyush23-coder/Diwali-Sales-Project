# Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Loading

```python
# Import CSV file

df=pd.read_csv("Diwali Sales Data.csv",
               encoding="unicode_escape")
df.head()
```

```python
# Check the sape of the Data

df.shape
```

# Ddatatypes of all column

```python
df.info()
```

```python
# Check Null Values 

df.isnull().sum()
```

# Clean the Data

```python
# Drop Null columns

df.drop(["Status","unnamed1"],
        axis=1,
        inplace=True)
df.info()
```

```python
# Drop Null Values Rows

df.dropna(inplace=True)
df.isnull().sum()
```

```python
# Change Datatype

df["Amount"] = df["Amount"].astype("int")
df["Amount"].dtypes
df.info()
```

```python
# Check All Column

df.columns
```

```python
# Rename the Column 

df.rename(columns=
          {"Cust_name":"Customer_Name"},
          inplace=True)
df.info()
```

# Description of the Data

```python
df.describe()
```

# Exploratory Data Analysis

## Gender

```python
df.groupby(["Gender"],as_index=False)["User_ID"].count()
```

```python
# Plotting a bar chart for Gender vs Amount

sales_gend = df.groupby(['Gender'],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Gender',y='Amount',data=sales_gend)
plt.show()
```

```python
# Plotting a bar chart for Gender vs it's count

px = sns.countplot(x='Gender',data=df)
sns.set(rc={'figure.figsize':(6,5)})
for bars in px.containers:
  px.bar_label(bars)
```

From above graphs we can see that most of the buyers ar females and even the purchasing power of females are greater then man

## Age

```python
# Plotting a bar chart for Amount vs Group

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age,palette="coolwarm")
```

From above graph we can see that most of the buyers are of age group between 26-35 Years

## Marital Status

```python
# Plotting a bar chart for Amount vs Marital Status

sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_state,palette='magma')
```

```python
# Plotting a bar chart for Marital Status vs it's count

ms=sns.countplot(x="Marital_Status",data=df)
for bar in ms.containers :
    ms.bar_label(bar)
```

From above graphs we can see that most of the buyers are married women and they have high purchasing power

## State

```python
# Plotting a bar chart for Order vs State

sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(10,5)})
# plt.Figure(figsize=(20,10))
sns.barplot(x='State',y='Orders',data=sales_state,palette='rocket')
```

```python
# Plotting a bar chart for Amount vs State

sales_state = df.groupby(['State'],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(10,5)})
ax=sns.barplot(x='State',y='Amount',data=sales_state,palette="viridis")
for bars in ax.containers:
 ax.bar_label(bars)
```

From above graphs we can see that most of the order and total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

## Occupation

```python
# Plotting a bar chart for Occupation and it's count

ax = sns.countplot(data = df, x = 'Occupation',palette="viridis")
plt.figure(figsize=(10,5))
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Plotting a bar chart for Amount vs Occupation

sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount',palette="viridis")
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation Sector

## Product Category

```python
# Plotting a bar chart for Procuct Catagory and it's count

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Plotting a bar chart for Product category vs Amount

sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

From above graphs we can see that most of the sold products are Food, Clothing, Electronics category

## Conclusion :

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category.
